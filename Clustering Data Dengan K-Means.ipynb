```python
from sklearn.datasets import make_blobs
import pandas as pd
```

```python
data, labels = make_blobs(n_samples=200, centers=3, cluster_std=0.5, random_state=42)
df = pd.DataFrame(data, columns=['Fitur 1', 'Fitur 2'])

print(df.head())
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df)
df['Cluster'] = kmeans.labels_
print(df['Cluster'].value_counts())
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Fitur 1'], df['Fitur 2'], c=df['Cluster'], cmap='viridis')
plt.title("Visualisasi Klastering Data dengan K-Means")
plt.xlabel("Fitur 1")
plt.ylabel("Fitur 2")
plt.show()
```
